```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Configuration
n_samples = 1000
priorities = ['very low', 'low', 'medium', 'high', 'very high']
assignees = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi']

# Seed for reproducibility
np.random.seed(42)
random.seed(42)

def generate_data(n):
    data = []

    for _ in range(n):
        assignee = random.choice(assignees)

        start_date = datetime.now() - timedelta(days=np.random.randint(0, 60))
        actual_duration_days = np.random.randint(1, 15)
        actual_duration = timedelta(days=actual_duration_days)
        completion_date = start_date + actual_duration

        est_error = random.uniform(-0.2, 0.2)  # Â±20% variation
        estimated_completion = start_date + timedelta(days=int(actual_duration_days * (1 + est_error)))

        task_priority = random.choice(priorities)
        remain_sub_task_level = random.choice(priorities)
        sub_task = f"{random.randint(1,5)} pending subtasks"
        total_assign_task = random.randint(1, 20)

        data.append({
            "assignee": assignee,
            "completion_date": completion_date,
            "estimated_completion_date": estimated_completion,
            "task_priority": task_priority,
            "sub_task": sub_task,
            "remain_sub_task_level": remain_sub_task_level,
            "actual_duration": actual_duration_days,
            "start_date": start_date,
            "total_assign_task": total_assign_task
        })

    return pd.DataFrame(data)

# Generate and save
df = generate_data(n_samples)
df.to_csv("task_data.csv", index=False)
print("âœ… CSV file 'task_data.csv' created successfully with assignees.")
```
